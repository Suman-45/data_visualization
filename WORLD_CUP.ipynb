```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv("WorldCupMatches.csv")
df=pd.DataFrame(data)
```

```python
#  i) no of rows and columns
print(len(data))
print(len(data.columns))
```

```python
# ii) print each column name
print(data.columns)
```

```python
# iii) print specific range using slicing
print(data[20:100])
```

```python
# iv) summarization statistics
print(data["Away Team Goals"].mean())
print(data["Away Team Goals"].median())
print(data["Away Team Goals"].mode()[0])
print(data["Away Team Goals"].max())
print(data["Away Team Goals"].min())
```

```python
#  i) check missing values
print(df.isnull())#show null values
print(df.isnull().sum())#print the sum of null values in a column
```

```python
# ii) if found in missing numerical values replace it by mean median
# df["Away Team Goals"].fillna(df["Away Team Goals"].mean(),inplace=True)
print(df)
df["Away Team Goals"].fillna(df["Away Team Goals"].median(),inplace=True)
print(df)
```
